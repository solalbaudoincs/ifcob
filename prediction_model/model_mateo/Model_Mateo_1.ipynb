```python
import numpy as np

target = df_eth["return-all-signed-for-5-ms"]

#pre_features = df_xbt[["return-all-signed-for-20-ms", "return-all-signed-for-10-ms", "return-all-signed-for-5-ms", "bid-ask-imbalance-5-levels", "spread"]]
pre_features = df_xbt[["bid-ask-imbalance-5-levels", "spread", "inst-return", "V-bid-5-levels", "V-ask-5-levels", "slope-bid-5-levels", "slope-ask-5-levels"]]

target_timestamp = target.index.values
feature_timestamp = pre_features.index.values
filtered_indicies = np.searchsorted(feature_timestamp, target_timestamp, side="left")

filtered_indicies = filtered_indicies[filtered_indicies < len(pre_features)]
pre_features_filtered = pre_features.iloc[filtered_indicies]

pre_features_filtered
```

```python
# Reset index of pre_features_filtered to match target's index
pre_features_aligned = pre_features_filtered.reset_index(drop=True)
pre_features_aligned.index = target.index

# Concatenate pre_features_aligned and target
combined_data = pd.concat([pre_features_aligned, target], axis=1)
combined_data.dropna(inplace=True)
combined_data
```

```python
target = combined_data["return-all-signed-for-5-ms"]
data = combined_data.drop(columns=["return-all-signed-for-5-ms"])
# Filter out rows where target equals 1
mask = target != 1
target = target[mask]
data = data[mask]

print(f"Data shape after filtering out class 1: {data.shape}")
print(f"Target shape after filtering out class 1: {target.shape}")
print("Remaining class distribution:")
print(target.value_counts().sort_index())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
from sklearn.ensemble import GradientBoostingClassifier
from lightgbm import LGBMClassifier

# Check class distribution
print("Class distribution in target:")
print(target.value_counts().sort_index())
print("\nClass proportions:")
print(target.value_counts(normalize=True).sort_index())

# Split the data into training and testing sets
# Temporal split - test set contains the last 20% of data chronologically
split_index = int(len(data) * 0.8)
X_train, X_test = data.iloc[:split_index], data.iloc[split_index:]
y_train, y_test = target.iloc[:split_index], target.iloc[split_index:]

# Calculate class weights to handle imbalance
class_weights = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
class_weight_dict = dict(zip(np.unique(y_train), class_weights))
print(f"\nClass weights: {class_weight_dict}")

# Create and train a small Random Forest model with balanced class weights
rf_model = RandomForestClassifier(
    n_estimators=100,  # Small number of trees for a quick test
    max_depth=3,  # Limit depth to prevent overfitting
    class_weight="balanced",
    random_state=42,
    n_jobs=-1  # Use all available cores
)
rf_model.fit(X_train, y_train)

# Make predictions
y_pred_rf = rf_model.predict(X_test)

# Evaluate the model
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"\nSmall Random Forest Accuracy: {accuracy_rf:.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_rf, labels=[-1, 0]))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf, labels=[-1, 0]))

# Feature importance
feature_importance = pd.DataFrame({
    'feature': data.columns,
    'importance': rf_model.feature_importances_
}).sort_values('importance', ascending=False)
print("\nFeature Importance:")
print(feature_importance)
#adjusted rand index
#from sklearn.metrics import adjusted_rand_score
#ari_score = adjusted_rand_score(y_test, y_pred_rf)
#ari_score
```

```python
from sklearn.dummy import DummyClassifier

# Create dummy classifiers with different strategies
dummy_most_frequent = DummyClassifier(strategy="most_frequent", random_state=42)
dummy_stratified = DummyClassifier(strategy="stratified", random_state=42)

# Fit dummy classifiers
dummy_most_frequent.fit(X_train, y_train)
dummy_stratified.fit(X_train, y_train)

# Make predictions
y_pred_dummy_freq = dummy_most_frequent.predict(X_test)
y_pred_dummy_strat = dummy_stratified.predict(X_test)

# Evaluate dummy classifiers
accuracy_dummy_freq = accuracy_score(y_test, y_pred_dummy_freq)
accuracy_dummy_strat = accuracy_score(y_test, y_pred_dummy_strat)

print("=== COMPARISON WITH DUMMY CLASSIFIERS ===")
print(f"Random Forest Accuracy: {accuracy_rf:.4f}")
print(f"Dummy (Most Frequent) Accuracy: {accuracy_dummy_freq:.4f}")
print(f"Dummy (Stratified) Accuracy: {accuracy_dummy_strat:.4f}")
print(f"\nImprovement over Most Frequent: {accuracy_rf - accuracy_dummy_freq:.4f}")
print(f"Improvement over Stratified: {accuracy_rf - accuracy_dummy_strat:.4f}")

print("\n=== DUMMY CLASSIFIER REPORTS ===")
print("\nDummy (Most Frequent) Classification Report:")
print(classification_report(y_test, y_pred_dummy_freq, labels=[-1, 0]))
print("\nDummy (Stratified) Classification Report:")
print(classification_report(y_test, y_pred_dummy_strat, labels=[-1, 0]))
```
