```python
import pandas as pd
import joblib
```

```python
df_eth = pd.read_parquet("../data/features/DATA_1/ETH_EUR.parquet")
df_xbt = pd.read_parquet("../data/features/DATA_1/XBT_EUR.parquet")
model = joblib.load("../predictors/random_forest_mateo_2/model-random_forest_mateo_2_/target-avg_10ms_of_mid_price_itincreases_after_200ms_with_threshold_5_depth-5_nest-100/model.joblib")
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1, figsize=(12, 6))

# Plot both datasets on the same axes
ax.plot(df_xbt["avg-250ms-of-V-bid-5-levels"], label='ETH avg-250ms-of-V-bid-5-levels', alpha=0.7)
ax.plot(df_xbt["V-bid-5-levels"], label='XBT V-bid-5-levels', alpha=0.7, color='orange')

ax.set_ylabel('Volume')
ax.set_xlabel('Timestamp')
ax.set_title('ETH avg-250ms-of-V-bid-5-levels vs XBT V-bid-5-levels')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
model_data = df_xbt[[
        "slope-bid-5-levels",
        "slope-ask-5-levels",
        "avg-250ms-of-slope-ask-5-levels",
        "avg-250ms-of-slope-bid-5-levels",
        "avg-250ms-of-V-bid-5-levels",
        "avg-250ms-of-V-ask-5-levels",
        "avg-250ms-of-liquidity-ratio-5-levels",
    ]]
predictions = model.predict(model_data)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 10), sharex=True)

# Plot predictions on the first subplot
ax1.plot(df_xbt.index, predictions, label='XBT Predictions', alpha=0.7, color='red')
ax1.set_ylabel('Predictions')
ax1.set_title('XBT Model Predictions of ETH')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot ETH data on the second subplot with dual y-axes
ax2_twin = ax2.twinx()

# Plot avg-10ms-of-mid-price on left y-axis
line1 = ax2.plot(df_eth.index, df_eth["avg-10ms-of-mid-price"], 
                 label='ETH avg-10ms-of-mid-price', alpha=0.7, color='blue')
ax2.set_ylabel('avg-10ms-of-mid-price', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Plot avg-10ms-of-mid-price-itincreases-after-200ms-with-threshold-5 on right y-axis
line2 = ax2_twin.plot(df_eth.index, df_eth["avg-10ms-of-mid-price-itincreases-after-200ms-with-threshold-5"], 
                      label='ETH target variable : avg-10ms-of-mid-price-itincreases-after-200ms-with-threshold-5', alpha=0.7, color='green')
ax2_twin.set_ylabel('avg-10ms-of-mid-price-itincreases-after-200ms-with-threshold-5', color='green')
ax2_twin.tick_params(axis='y', labelcolor='green')

# Combine legends
lines1, labels1 = ax2.get_legend_handles_labels()
lines2, labels2 = ax2_twin.get_legend_handles_labels()
ax2.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

ax2.set_xlabel('Timestamp')
ax2.set_title('ETH: Price vs Target Variable')
ax2.grid(True, alpha=0.3)


plt.tight_layout()
plt.show()
```

```python
# Create aligned data by reindexing to match timesteps
# Use the XBT dataframe index as the reference since that's what the model was trained on
aligned_eth_target = df_eth["avg-10ms-of-mid-price-itincreases-after-200ms-with-threshold-5"].reindex(df_xbt.index, method='nearest')

import numpy as np

# Define number of bins
n_bins = 5000

# Create bins for the data
bins = np.linspace(0, len(aligned_eth_target)-1, n_bins+1, dtype=int)

# Initialize arrays to store binned data
binned_target = []
binned_predictions = []
binned_indices = []

# Calculate average values for each bin
for i in range(len(bins)-1):
    start_idx = bins[i]
    end_idx = bins[i+1]
    
    # Calculate average target value for this bin
    target_avg = aligned_eth_target.iloc[start_idx:end_idx].sum()
    
    # Calculate average prediction value for this bin
    pred_avg = predictions[start_idx:end_idx].sum()
    
    # Use middle index for x-axis
    mid_idx = (start_idx + end_idx) // 2
    
    binned_target.append(target_avg)
    binned_predictions.append(pred_avg)
    binned_indices.append(df_xbt.index[mid_idx])

# Create the plot with binned data
fig, ax = plt.subplots(1, 1, figsize=(15, 8))

# Plot binned data
ax.plot(binned_indices, binned_target, label=f'ETH Target (Binned - {n_bins} bins)', alpha=0.8, color='blue', marker='o', markersize=3)
ax.plot(binned_indices, binned_predictions, label=f'XBT Model Predictions (Binned - {n_bins} bins)', alpha=0.8, color='red', marker='s', markersize=3)

ax.set_ylabel('Average Target/Prediction Values')
ax.set_xlabel('Timestamp')
ax.set_title(f'Model Target vs Model Predictions (Binned Data - {n_bins} bins)')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"Original data points: {len(aligned_eth_target)}")
print(f"Binned data points: {len(binned_target)}")
print(f"Data reduction: {len(aligned_eth_target)/len(binned_target):.1f}x")
```
