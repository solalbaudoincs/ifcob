```python
import pandas as pd
import joblib
```

```python
df_eth = pd.read_parquet("../data/features/DATA_0/ETH_EUR.parquet")
df_xbt = pd.read_parquet("../data/features/DATA_0/XBT_EUR.parquet")
model = joblib.load("../predictors/mateo/target-avg_10ms_of_mid_price_itincreases_after_200ms_with_threshold_5_depth-5_nest-100/model.joblib")
```

```python
df_eth
```

```python
df_xbt
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1, figsize=(12, 6))

# Plot both datasets on the same axes
ax.plot(df_xbt["avg-250ms-of-V-bid-5-levels"], label='ETH avg-250ms-of-V-bid-5-levels', alpha=0.7)
ax.plot(df_xbt["V-bid-5-levels"], label='XBT V-bid-5-levels', alpha=0.7, color='orange')

ax.set_ylabel('Volume')
ax.set_xlabel('Timestamp')
ax.set_title('ETH avg-250ms-of-V-bid-5-levels vs XBT V-bid-5-levels')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
model_data = df_xbt[[
        "slope-bid-5-levels",
        "slope-ask-5-levels",
        "avg-250ms-of-slope-ask-5-levels",
        "avg-250ms-of-slope-bid-5-levels",
        "avg-250ms-of-V-bid-5-levels",
        "avg-250ms-of-V-ask-5-levels",
        "avg-250ms-of-liquidity-ratio-5-levels",
    ]]
predictions = model.predict(model_data)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 10), sharex=True)

# Plot predictions on the first subplot
ax1.plot(df_xbt.index, predictions, label='XBT Predictions', alpha=0.7, color='red')
ax1.set_ylabel('Predictions')
ax1.set_title('XBT Model Predictions of ETH')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot ETH data on the second subplot with dual y-axes
ax2_twin = ax2.twinx()

# Plot avg-10ms-of-mid-price on left y-axis
line1 = ax2.plot(df_eth.index, df_eth["avg-10ms-of-mid-price"], 
                 label='ETH avg-10ms-of-mid-price', alpha=0.7, color='blue')
ax2.set_ylabel('avg-10ms-of-mid-price', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Plot avg-10ms-of-mid-price-itincreases-after-200ms-with-threshold-5 on right y-axis
line2 = ax2_twin.plot(df_eth.index, df_eth["avg-10ms-of-mid-price-itincreases-after-200ms-with-threshold-5"], 
                      label='ETH target variable : avg-10ms-of-mid-price-itincreases-after-200ms-with-threshold-5', alpha=0.7, color='green')
ax2_twin.set_ylabel('avg-10ms-of-mid-price-itincreases-after-200ms-with-threshold-5', color='green')
ax2_twin.tick_params(axis='y', labelcolor='green')

# Combine legends
lines1, labels1 = ax2.get_legend_handles_labels()
lines2, labels2 = ax2_twin.get_legend_handles_labels()
ax2.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

ax2.set_xlabel('Timestamp')
ax2.set_title('ETH: Price vs Target Variable')
ax2.grid(True, alpha=0.3)


plt.tight_layout()
plt.show()
```
